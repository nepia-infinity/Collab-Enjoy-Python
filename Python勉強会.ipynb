# スクレイピング

スクレイピングは違法の場合もある。
利用規約などを事前に読み込む必要がある
会社で実務の一環で行う場合は、会社の品位を疑われるケースもあるので気をつけよう

https://portal.data.metro.tokyo.lg.jp/terms/


```python
import requests
from bs4 import BeautifulSoup
from google.colab import files
import time

# スクレイピング対象のURL
url = "https://portal.data.metro.tokyo.lg.jp/"
# GETリクエストを送信してHTMLを取得
response = requests.get(url)

# BeautifulSoupでHTMLを解析
soup = BeautifulSoup(response.content, "html.parser")

# タイトルを取得
title = soup.find("title").text
print("タイトル:", title)

# imgタグを全て取得する
elements = soup.find_all('img')

for i, element in enumerate(elements, start=1):
    # src属性を取得 (存在しない場合は空文字列)
    src = element.get('src', '')

    if not src.startswith('https'):
        src = url + src
    print(src)

    # f-string と文字列フォーマットでファイル名を生成
    filename = f'東京都オープンデータカタログ_{i:03d}.jpg'

    # 画像をダウンロードして保存(いくつか保存に失敗する)
    image_response = requests.get(src) # URLから画像コンテンツを取得
    if image_response.status_code == 200:
        with open(filename, 'wb') as f:
            f.write(image_response.content) # 画像コンテンツをファイルに保存
        print(f"Downloaded and saved {filename}")
        files.download(filename) # 保存したファイルをダウンロード
    else:
        print(f"Failed to download image from {src}")

    # リクエスト間に遅延を入れる
    time.sleep(1)
```

```python
import requests
from bs4 import BeautifulSoup

from google.colab import auth
auth.authenticate_user()

# Google Spreadsheetとの連携
import gspread
from google.auth import default
creds, _ = default()
gc = gspread.authorize(creds)


def get_article():

  url = "https://snabi.jp/article"
  response = requests.get(url)
  response.raise_for_status()
  soup = BeautifulSoup(response.content, 'html.parser')

  # class="article-card__title" を持つ div タグを全て取得
  title_divs = soup.find_all('div', class_='article-card__title')
  article_link = soup.find_all('a', class_='article-card__link' )

  values = [['title', 'url']]

  # zip を使って title_divs と article_link を同時にループ処理
  for i, (title_div, link) in enumerate(zip(title_divs, article_link)):
      article_number = i + 1
      title = title_div.text.strip()
      href = link.get('href')
      link  = f"https://snabi.jp{href}"

      values.append([title, link])

      print(f"{article_number}. {title}")
      print(link)

  print(values)
  return values



def main(values):
  sheet_url = "https://docs.google.com/spreadsheets/d/************"
  spreadsheet = gc.open_by_url(sheet_url)
  sheet = spreadsheet.worksheet("記事一覧")
  print(f'\nファイル名：{spreadsheet.title}')
  print(f'シート名：{sheet.title}')

  sheet.update("A1", values)

  print("Data written to Google Sheets successfully!")
  print(sheet_url)
  return



if __name__ == "__main__":
    values = get_article()
    main(values)
```

# PDFの操作（内容の取得・分割・結合）

応用としてPDFの全文を取得、Open AIに投げて要約をするとかもアリかも

```python
import PyPDF2
from google.colab import files

def split_pdf(uploaded_pdf, output_pdf_prefix):
    """
    PDFファイルを分割する関数

    Args:
        uploaded_pdf (dict): アップロードされたPDFファイルの情報を含む辞書
        output_pdf_prefix (str): 出力PDFファイルの接頭辞
    """

    # アップロードされたファイル名を取得
    input_pdf_filename = list(uploaded_pdf.keys())[0]

    # ファイル名から拡張子を除いた部分を取得
    base_filename = input_pdf_filename.split('.')[0]

    with open(input_pdf_filename, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(pdf_reader.pages)

        for page_num in range(num_pages):
            pdf_writer = PyPDF2.PdfWriter()
            pdf_writer.add_page(pdf_reader.pages[page_num])

            # アップロードされたファイル名とページ番号を組み合わせた出力ファイル名を生成
            output_filename = f"{base_filename}_{page_num + 1:03d}.pdf"
            with open(output_filename, 'wb') as output_pdf:
                pdf_writer.write(output_pdf)
                files.download(output_filename)


if __name__ == "__main__":
    uploaded_pdf = files.upload()
    output_pdf_prefix = 'split_pdf'  # 出力ファイルの接頭辞を指定

    split_pdf(uploaded_pdf, output_pdf_prefix)
```

# 画像の編集

```python
from google.colab import drive
import PIL.Image as Image
import os
from datetime import datetime
import pytz

# Google Driveをマウント
drive.mount('/content/drive')

# 画像が保存されているフォルダのパス
folder_path = '/content/drive/MyDrive/original_images'  # 適宜変更してください

# 実行日時を取得し、タイムゾーンを Asia/Tokyo に設定
tokyo_timezone = pytz.timezone('Asia/Tokyo')
now = datetime.now(tokyo_timezone)

# フォルダ名に使用する文字列を生成 (例: 1205_031)
timestamp = now.strftime("%m%d_%H%M%s")

# リサイズ後の画像を保存するフォルダのパス (実行日時を含む)
output_folder_path = f'/content/drive/MyDrive/original_images/resized_{timestamp}'

# Google Drive 上にフォルダを作成 (osモジュールを使用)
os.mkdir(output_folder_path)

# 画像ファイルの拡張子
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.gif']  # 拡張子を追加

# 指定されたフォルダ内の画像ファイルを処理
for filename in os.listdir(folder_path):
    # ファイルパスを作成
    file_path = os.path.join(folder_path, filename)

    # ファイルが画像ファイルであるか確認
    if os.path.isfile(file_path) and any(filename.lower().endswith(ext) for ext in image_extensions):

      try:
          # 画像を開く
          img = Image.open(file_path)

          # アスペクト比を維持したまま、最大サイズを (512, 512) にリサイズ
          img.thumbnail((512, 512))  # 戻り値は None なので代入しない

          # リサイズした画像を保存 (元の拡張子を維持)
          new_file_name = f"{os.path.splitext(filename)[0]}_resized{os.path.splitext(filename)[1]}"
          output_file_path = os.path.join(output_folder_path, new_file_name)
          img.save(output_file_path)

          print(f"{new_file_name}として保存しました。")

      except Image.UnidentifiedImageError:
          print(f"Error: {new_file_name} の処理は途中で中断されました")

print("すべてのファイルの保存が完了しました")
```

```python
import requests, json
from google.colab import userdata

# Google Spreadsheetとの連携
import gspread
from google.auth import default
creds, _ = default()
gc = gspread.authorize(creds)

def main():
  # 地区コードAPIのURL
  request_url = "https://app.rakuten.co.jp/services/api/Travel/GetAreaClass/20131024"
  app_id = userdata.get("RAKUTEN_APP_ID")

  params = {"format": "json", "applicationId": app_id}

  # 地区コードの取得
  response = requests.get(request_url, params=params)
  if response.status_code == 200:
    area_data = response.json()
    # JSONデータをインデント付きで表示
    print(json.dumps(area_data, indent=4, ensure_ascii=False))

    # Google Spreadsheetにデータを保存
    set_area_data_to_spreadsheet(area_data)


  else:
    print(f"Error: API request failed with status code {response.status_code}")



def convert_to_2d_array(data):
  result = [["middleClassCode", "middleClassName", "smallClassCode", "smallClassName"]]
  middles = data["areaClasses"]["largeClasses"][0]["largeClass"][1]["middleClasses"]
  # print(middles)

  for middle in middles:
      print(middle)

      for small in middle["middleClass"][1]["smallClasses"]:
        print(f"★{small}")

        middle_code = middle["middleClass"][0]["middleClassCode"]
        middle_name = middle["middleClass"][0]["middleClassName"]
        small_code = small["smallClass"][0]["smallClassCode"]
        small_name = small["smallClass"][0]["smallClassName"]

        result.append([middle_code, middle_name, small_code, small_name])

  # 結果を2次元配列として返す
  print(result)
  return result



def set_area_data_to_spreadsheet(area_data):
  sheet_url = "https://docs.google.com/spreadsheets/d/12xchnCScSVOyySm6wgTiBw3V0720vXBFtPxX-Wzo4HM/edit?gid=175127849#gid=175127849"
  spreadsheet = gc.open_by_url(sheet_url)
  sheet = spreadsheet.worksheet("楽天ホテル検索API")
  print(f'\nファイル名：{spreadsheet.title}')
  print(f'シート名：{sheet.title}')

  values = convert_to_2d_array(area_data)
  sheet.update("A1", values)

  print("Data written to Google Sheets successfully!")
  print(sheet_url)
  return


if __name__ == "__main__":
    main()
```
